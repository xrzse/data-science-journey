## Do Hungrier People Spend More?

```python
hungyppl = df[["hunger_level","order_value"]]
hungyppl
```

```python
print(hungyppl["hunger_level"].value_counts())
```

```python
print(hungyppl.groupby("hunger_level")["order_value"].mean()) #groupby, chooses all rows in hunger_level and pairs up w ordervalue
```

```python
hungyppl.groupby("hunger_level")["order_value"].agg(["mean", "std", "min", "max"]) #agg allows to do more functions at once
```

Therefore hunger level does not affect how much is spent on food
